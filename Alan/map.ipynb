```python
import os
import warnings
os.chdir("..")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
```

```python
# 1. Load your clustered data
df = pd.read_csv('data/combined_csvs/listings_clustered.csv')

# 2. Filter to a single city to see the neighborhood clusters clearly
city_to_map = 'asheville'
df_city = df[df['city'] == city_to_map]

# 3. Create the interactive map
fig = px.scatter_mapbox(
    df_city,
    lat="latitude",
    lon="longitude",
    color="cluster_id",                  # This automatically color-codes your clusters!
    hover_name="listing_id",             # What shows up in bold when you hover
    hover_data=["ttm_revenue", "gross_yield", "cluster_avg_ttm_revenue"], # Extra hover info
    zoom=11,                             # Starting zoom level (10-12 is good for cities)
    title=f"Airbnb Market Clusters in {city_to_map.title()}"
)

# 4. Set the map style to a free, built-in map (no API keys needed)
fig.update_layout(
    mapbox_style="open-street-map",
    margin={"r":0,"t":40,"l":0,"b":0}    # Removes wasted white space around the edges
)

# 5. Launch the map in your web browser
fig.show()
```

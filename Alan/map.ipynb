```python
import pandas as pd
import plotly.express as px

# 1. Load data
df = pd.read_csv("data/combined_csvs/listings_clustered.csv")

# 2. Filter city
df_city = df[df["city"] == city_to_map].copy()

# Compute mean gross_yield per cluster and map back
cluster_yields = df_city.groupby("cluster_id")["gross_yield"].transform("mean")
df_city["cluster_avg_gross_yield"] = cluster_yields

# 3. Plot clusters heatmapped by average gross yield
fig1 = px.scatter_mapbox(
    df_city,
    lat="latitude",
    lon="longitude",
    color="cluster_avg_gross_yield",  # Color by cluster's average yield
    color_continuous_scale="Viridis",  # Heatmap color scale
    hover_name="listing_id",
    hover_data=[
        "cluster_id",
        "gross_yield",
        "cluster_median_gross_yield",
        "ttm_revenue",
        "property_val",
    ],
    zoom=11,
    title=f"Airbnb Market Clusters in {city_to_map.title()} (Heatmapped by Cluster Avg Gross Yield)",
    width=1000,
    height=600,
)

fig1.update_layout(
    mapbox_style="carto-positron", margin={"r": 0, "t": 40, "l": 0, "b": 0}
)

fig1.show()
```

```python
# Plot individual dots heatmapped with fixed color range (0 to 0.2)
fig2 = px.scatter_mapbox(
    df_city,
    lat="latitude",
    lon="longitude",
    color="gross_yield",
    range_color=[0, 0.20],  # Caps color scale at 0.2 so 0.4 doesn't skew everything
    color_continuous_scale="Viridis",
    hover_name="listing_id",
    hover_data=["gross_yield", "ttm_revenue", "cluster_id", "property_val"],
    zoom=11,
    title=f"Airbnb Listings in {city_to_map.title()} (Individual Gross Yield Heatmap)",
    width=1000,
    height=600,
)

fig2.update_layout(
    mapbox_style="carto-positron", margin={"r": 0, "t": 40, "l": 0, "b": 0}
)

fig2.show()
```

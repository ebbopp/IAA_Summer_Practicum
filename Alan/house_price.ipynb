```python
zillow_dfs = []
for n_bedrooms in range(1, 6):
    file_path = f"/Users/alanchen/Documents/IAA_Summer_Practicum/data/zillow/Zip_zhvi_bdrmcnt_{n_bedrooms}_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv"
    df = pd.read_csv(file_path)
    df = df.copy()
    df["bedrooms"] = n_bedrooms
    zillow_dfs.append(df)

# Now combine them all at once
zillow_df = pd.concat(zillow_dfs, ignore_index=True)
```

```python
# Filter to private rooms and fill the missing values in 'bedrooms' as 1
df_listings.loc[df_listings['room_type'] == 'private_room', 'bedrooms'] = df_listings.loc[df_listings['room_type'] == 'private_room', 'bedrooms'].fillna(1)
# Filter to shared rooms and fill the missing values in 'bedrooms' as 1
df_listings.loc[df_listings['room_type'] == 'shared_room', 'bedrooms'] = df_listings.loc[df_listings['room_type'] == 'shared_room', 'bedrooms'].fillna(1)
```

```python
# fill remaining na value for bedrooms with the median value for the corresponding number of guests 
df_listings['bedrooms'] = df_listings['bedrooms'].fillna(
    df_listings.groupby(['guests'])['bedrooms'].transform('median')
)
```

```python
df_listings[df_listings["bedrooms"].isna()]['room_type'].value_counts()
```

```python
#remove extra columns from geocodio
cols_to_remove = [col for col in df_listings.columns if ("Geocodio" in col) and (col != "Geocodio Address Line 1")]
df_listings = df_listings.drop(columns=cols_to_remove)
```

```python
# Match on BOTH zipcode and bedrooms to keep it a clean 1:1 row match
df_listings = df_listings.merge(
    zillow_df[["zipcode", "bedrooms", "2026-05-31", "SizeRank"]], 
    how="left", 
    on=["zipcode", "bedrooms"]
).rename(columns={"2026-05-31": "property_val"})
```

```python
df_listings["gross_yield"] = df_listings["ttm_revenue"] / df_listings["property_val"]
```
